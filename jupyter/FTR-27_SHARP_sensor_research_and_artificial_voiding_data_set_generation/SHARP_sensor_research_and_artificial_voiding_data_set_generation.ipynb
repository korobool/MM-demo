````python
N = len(files)
fig, axes = plt.subplots(1, N) #sharey=True
fig.set_size_inches(6*N, 6) 
plt.tight_layout()

k_diversity = []

for i, d in enumerate(nD):
    # X = d['time_ms']
    # w = d['weight(g)']
    # f = d['flow_meter']
   

    df = d[d['valve_state'] == 1].reset_index(drop=True)
    df['k'] = df['flow_meter']/ (df['weight(g)'] - df['weight(g)'][0])
    X = df['time_ms']
    w = df['weight(g)']
    f = df['flow_meter']
    k = df['k']
    l= len(k)

    k_end = round(k[l-1],3)
    k_diversity.append(k_end)

    axes[i].plot(X, w, label = f' {i}: {files[i]}\n w=b  f=r  k=o  w(f)=g \n k-end={k_end}', linewidth=2.0, c='blue')
    axes[i].legend(loc='lower right')
    axes[i].twinx().plot(X, f, label = 'flow', linewidth=2.0, c='red')    
    # axes[i].plot(X, f, label = 'flow', linewidth=2.0, c='red') 

    axes[i].twinx().plot(X, k, label = 'k', linewidth=2.0, c='orange')
    axes[i].twinx().twiny().plot(f, w, label = 'weight', linewidth=2.0, c='green')    
    

    # axes[i].set_ylabel('distance(mm)')
    axes[i].set_xlabel('time(ms)')
plt.show()

print('kmin {} \nkmax {}\naccuracy {}'.format(np.min(k_diversity), np.max(k_diversity), np.min(k_diversity)/np.max(k_diversity)))
````
